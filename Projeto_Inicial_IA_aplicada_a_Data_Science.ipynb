### **2º Questionamento:** Qual foi o faturamento por categoria em 2023?

**Observação:** Para o faturamento, vamos considerar aqui o produto entre o preço unitário e a quantidade somadas ao frete de cada produto.

> #### <strong><font color="7b71ff" size =6>Prompt:</font></strong>

Temos um DataFrame chamado "df" e queremos construir um visual de gráfico de barras horizontais que apresente o faturamento por categoria da Zoop.

Em "df", temos as colunas:
* categoria: categorias dos produtos
* preco_unitario: preço de uma unidade do produto
* quantidade: quantidade vendida do produto
* frete: frete a ser pago no total da compra.

Desejamos criar uma visualização que apresente o faturamento de cada categoria, em que o faturamento é dado pelo produto entre o preço unitário e a quantidade somadas ao frete.

Utilize as bibliotecas de visualização de dados do Python para criar essa visualização.

````python
# Criando uma nova coluna com o faturamento total de cada linha
df['faturamento'] = (df['preco_unitario'] * df['quantidade']) + df['frete']

# Agrupando por categoria e somando o faturamento
faturamento_por_categoria = df.groupby('categoria')['faturamento'].sum().sort_values()

# Criando o gráfico de barras horizontais
plt.figure(figsize=(10, 6))
faturamento_por_categoria.plot(kind='barh', color='skyblue')

plt.title('Faturamento por Categoria (Zoop)', fontsize=14)
plt.xlabel('Faturamento (R$)', fontsize=12)
plt.ylabel('Categoria', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
````

<font color="7b71ff" size=4> Comentários: </font>

---

Aqui especificamos o tipo de gráfico, guiando o ChatGPT para criação do nosso visual. Podemos notar que em um gráfico de barras o foco é maior nas categorias, além de ser uma forma mais fluida de ler textos que possam ser extensos demais para um gráfico de coluna.

### **3º Questionamento:** Como foram as vendas ao longo dos meses em 2023?

**Observação:** É importante especificar como a sua variável tempo está especificada dentro da base para o ChatGPT alcançar resultados melhores na sugestão do visual

> #### <strong><font color="7b71ff" size =6>Prompt:</font></strong>

Temos um DataFrame chamado "df" e queremos construir um visual de gráfico de linha que apresente as vendas totais mensais da Zoop.

Em "df", temos as colunas:
* data: com as datas das compras no formato datetime (`aaaa-mm-dd`)
* faturamento: dado pelo produto entre o preço unitário e a quantidade somadas ao frete de cada venda.

Desejamos criar uma visualização que apresente as vendas por mês. Prmeiro, agrupe os dados por mês e depois crie uma nova coluna chamada "mes" que receba o nome de cada mês, traduzindo os meses do índice por meio do dicionário abaixo. Utilize a coluna "data" para o agrupamento dos dados e construção do gráfico. A coluna "mes" só deve ser utilizada para alterar os rótulos do eixo x.

```
meses = {
    'January': 'Jan',
    'February': 'Fev',
    'March': 'Mar',
    'April': 'Abr',
    'May': 'Mai',
    'June': 'Jun',
    'July': 'Jul',
    'August': 'Ago',
    'September': 'Set',
    'October': 'Out',
    'November': 'Nov',
    'December': 'Dez'
}
```

Utilize as bibliotecas de visualização de dados do Python para criar essa visualização.

````python
# Dicionário de tradução dos meses
meses = {
    'January': 'Jan',
    'February': 'Fev',
    'March': 'Mar',
    'April': 'Abr',
    'May': 'Mai',
    'June': 'Jun',
    'July': 'Jul',
    'August': 'Ago',
    'September': 'Set',
    'October': 'Out',
    'November': 'Nov',
    'December': 'Dez'
}

# Garantindo que a coluna "data" esteja no formato datetime
df['data'] = pd.to_datetime(df['data'])

# Agrupando os dados por mês e somando o faturamento
df_mensal = df.resample('M', on='data')['faturamento'].sum().reset_index()

# Criando a coluna com o nome do mês traduzido
df_mensal['mes'] = df_mensal['data'].dt.strftime('%B').map(meses)

# Criando o gráfico de linha
plt.figure(figsize=(9, 5))
plt.plot(df_mensal['mes'], df_mensal['faturamento'], marker='o', linestyle='-', color='teal')

plt.title('Vendas Totais Mensais da Zoop', fontsize=14)
plt.xlabel('Mês', fontsize=12)
plt.ylabel('Faturamento (R$)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
````

<font color="7b71ff" size=4> Comentários: </font>

---

Quando possuimos uma quantidade de períodos maior para os nossos dados, vale a pena recorrer a gráficos de linha(s) que é ideal para séries temporais. Como nossos dados são mensais, utilizar um gráfico de colunas, por exemplo, reduziria a percepção de uma sequência dos dados. Linhas dão uma maior sensação de continuidade.

Temos um DataFrame chamado "df" e queremos construir um visual de gráfico de barras empilhadas que apresente as vendas por trimestre em relação ao método de pagamento da Zoop.

Em "df", temos as colunas:
* data: com as datas das compras no formato datetime (`aaaa-mm-dd`)
* faturamento: com os valores das vendas
* metodo_pagamento: com o método de pagamento escolhido na compra

Desejamos criar uma visualização que apresente as vendas por trimestre, agrupando os métodos de pagamento de cada trimestre correspondente a cada compra.

Utilize as bibliotecas de visualização de dados do Python para criar essa visualização.

````python
# 1. Converter a coluna 'data' para datetime (caso ainda não esteja)
df['data'] = pd.to_datetime(df['data'])

# 2. Criar uma nova coluna para o trimestre
df['trimestre'] = df['data'].dt.to_period('Q')

# 3. Agrupar por trimestre e método de pagamento e somar o faturamento
vendas_trimestre = df.groupby(['trimestre', 'metodo_pagamento'])['faturamento'].sum().unstack(fill_value=0)

# 4. Plotar gráfico de barras empilhadas
vendas_trimestre.plot(kind='bar', stacked=True, figsize=(10,6), colormap='viridis')

# 5. Ajustes estéticos
plt.title('Faturamento por Trimestre e Método de Pagamento')
plt.xlabel('Trimestre')
plt.ylabel('Faturamento (R$)')
plt.xticks(rotation=45)
plt.legend(title='Método de Pagamento')
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)

# 6. Mostrar gráfico
plt.show()
````

````python
# Obter percentual que cada método representou no trimestre
percentual_trimestre = vendas_trimestre.div(vendas_trimestre.sum(axis=1), axis=0) * 100

# Plotar gráfico de barras empilhadas com percentuais
percentual_trimestre.plot(kind='bar', stacked=True, figsize=(10,6), colormap='Set2')

plt.title('Distribuição Percentual dos Métodos de Pagamento por Trimestre')
plt.ylabel('Percentual (%)')
plt.xlabel('Trimestre')
plt.xticks(rotation=45)
plt.legend(title='Método de Pagamento', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
````

````python

def vendas_tri_metodo():
    # Converter a coluna 'data' para o formato datetime, se ainda não estiver
    df['data'] = pd.to_datetime(df['data'])

    # Criar uma nova coluna para identificar o trimestre
    df['trimestre'] = df['data'].dt.to_period('Q').astype(str)

    # Agrupar por trimestre e método de pagamento, somando o faturamento
    df_grouped = df.groupby(['trimestre', 'metodo_pagamento'])['faturamento'].sum().reset_index()

    # Pivot para gráfico empilhado
    pivot_df = df_grouped.pivot(index='trimestre', columns='metodo_pagamento', values='faturamento').fillna(0)
    pivot_df = pivot_df.sort_index()  # garantir ordem dos trimestres

    # Cores para os métodos (ordem conforme as colunas)
    palette = [BRANCO, CINZA_3, AQUA_1, AQUA_3]
    colors = palette[:len(pivot_df.columns)]

    # Criar figura
    fig, ax = plt.subplots(figsize=(10, 6))
    fig.patch.set_facecolor(CINZA_1)
    ax.set_facecolor(CINZA_1)

    # Plotar gráfico de barras empilhadas
    bottom = None
    for i, col in enumerate(pivot_df.columns):
        ax.bar(
            pivot_df.index,
            pivot_df[col],
            bottom=bottom,
            label=col,
            color=colors[i]
        )
        bottom = pivot_df[col] if bottom is None else bottom + pivot_df[col]

    # Título
    ax.set_title('Vendas por Trimestre e Método de Pagamento em 2023', fontsize=18, color=CINZA_5)

    # Eixo X
    ax.tick_params(axis='x', labelsize=12, colors=AZUL_1, rotation=0)
    ax.set_xlabel('')  # remove label eixo x

    # Eixo Y
    ax.set_ylabel('')
    # Get current y-tick locations before setting custom labels
    current_yticks = ax.get_yticks()
    yticks_labels = [f'R$ {int(y/1e6)} M' for y in current_yticks]
    ax.set_yticklabels(yticks_labels, fontsize=12, color=AZUL_1)


    # Remover bordas
    for spine in ['top', 'bottom', 'left', 'right']:
        ax.spines[spine].set_visible(False)

    # Grid em Y (tracejado)
    ax.yaxis.grid(True, linestyle='--', linewidth=0.7, alpha=0.6)
    ax.xaxis.grid(False)

    # Legenda (opcional)
    ax.legend(title='Método de Pagamento', fontsize=10)

    return plt # Changed return to return plt
````

````python
def metodos_pag():
  # Agrupar por método de pagamento e contar o total de ocorrências
  metodos_de_pagamento = df['metodo_pagamento'].value_counts()

  # Calcular percentual
  total_clientes = metodos_de_pagamento.sum()
  percentual = (metodos_de_pagamento[0] + metodos_de_pagamento[1]) / total_clientes * 100

  # Configurações do gráfico
  plt.figure(figsize=(10, 6), facecolor=CINZA_1)
  ax = plt.axes()
  ax.set_facecolor(CINZA_1)
  plt.bar(metodos_de_pagamento.index, metodos_de_pagamento, color=VERMELHO_1)

  # Adicionar texto com o total acima de cada barra
  for i, v in enumerate(metodos_de_pagamento):
      plt.text(i, v + 0.1, str(v), ha='center', va='bottom', color=AZUL_1, fontsize=12)

  # Adicionar texto dentro da área do plot
  plt.text(2, 2500, f"$\\bf{percentual:.2f}$% dos clientes utilizam Cartão de Crédito ou PIX \npara pagamento. Seria interessante recorrer a parcerias\ncom bancos para a construção de um Zoop Pay a fim\nde fidelizar a nossa clientela.",
          ha='left', va='center', color=AZUL_3, fontsize=10)

  # Configurações adicionais
  plt.title('Métodos de Pagamentos mais Utilizados em 2023', fontsize=18, color=CINZA_5)
  plt.xticks(fontsize=12, color=AZUL_1)
  plt.yticks([])
  plt.xlabel('Método de Pagamento', fontsize=12, color=AZUL_1)
  ax.spines[['top', 'right', 'left']].set_visible(False)
  plt.grid(False)

  return plt

def vendas_mensais():
  # Dicionário de tradução de meses
  meses = {'January': 'Jan', 'February': 'Fev', 'March': 'Mar', 'April': 'Abr',
           'May': 'Mai', 'June': 'Jun', 'July': 'Jul', 'August': 'Ago',
           'September': 'Set', 'October': 'Out', 'November': 'Nov', 'December': 'Dez'}

  # Agrupar por mês e calcular as vendas totais
  df['mes'] = df['data'].dt.strftime('%B')
  vendas_mensais = df.groupby('mes')['faturamento'].sum().reindex(meses.keys())

  # Configurações do gráfico
  plt.figure(figsize=(12, 6), facecolor=CINZA_1)
  ax = plt.axes()
  ax.set_facecolor(CINZA_1)
  plt.plot(vendas_mensais.index, vendas_mensais, marker='o', color=AZUL_1, linewidth=2)

  # Adicionar rótulos e título
  plt.title('Vendas Totais Mensais em 2023', fontsize=18, color=CINZA_5)
  plt.xlabel('')
  plt.ylabel('')

  # Formatando rótulos do eixo y em "R$ valor M"
  formatted_yticks = ['R$ {:.2f} M'.format(val / 1e6) for val in ax.get_yticks()]
  ax.set_yticks(ax.get_yticks()) # para evitar o Warning
  ax.set_yticklabels(formatted_yticks, fontsize=12, color=AZUL_1)

  # Adicionar círculos numerados
  for i, point in enumerate([(4, vendas_mensais.iloc[4]), (7, vendas_mensais.iloc[7]), (10, vendas_mensais.iloc[10]), (11, vendas_mensais.iloc[11])], start=1):
      plt.scatter(point[0], point[1], s=200, color='none', edgecolors=VERMELHO_1, alpha=0.7)
      plt.text(point[0] - 0.3, point[1] + 0.1, str(i), fontsize=12, color=VERMELHO_1)


  # Configurações adicionais
  plt.xticks(vendas_mensais.index, [meses[mes] for mes in vendas_mensais.index], fontsize=12, color=AZUL_1)
  ax.spines['top'].set_visible(False)
  ax.spines['right'].set_visible(False)


  return plt

def faturamento_por_categoria():
    # Calculate the faturamento for each category
    # This calculation is already present in a previous cell (dGthrTKuejNz) and should be part of the 'df' DataFrame.
    # If 'faturamento' is not in the 'df' DataFrame, add the calculation here:
    # df['faturamento'] = (df['preco_unitario'] * df['quantidade']) + df['frete']

    # Agrupar por categoria e calcular o faturamento total
    faturamento_por_categoria = df.groupby('categoria')['faturamento'].sum().reset_index()

    # Ordenar por faturamento
    faturamento_por_categoria = faturamento_por_categoria.sort_values(by='faturamento', ascending=True)

    # Criar a figura
    fig, ax = plt.subplots(figsize=(10, 6)) # Changed figure size to match prompt
    fig.patch.set_facecolor(CINZA_1)
    ax.set_facecolor(CINZA_1)

    # Plotar barras horizontais
    bars = ax.barh(faturamento_por_categoria['categoria'], faturamento_por_categoria['faturamento'], color=AQUA_1)

    # Título
    ax.set_title('Faturamento por categoria de produto em 2023', fontsize=18, color=CINZA_5)

    # Remover labels dos eixos e xticks
    ax.set_xlabel('')
    ax.set_ylabel('')
    ax.set_xticks([])

    # Estilizar yticks
    ax.tick_params(axis='y', labelsize=12, colors=AZUL_1)

    # Remover todas as bordas
    for spine in ax.spines.values():
        spine.set_visible(False)

    # Adicionar os valores ao lado direito das barras
    for bar in bars:
        ax.text(bar.get_width() + 100000, bar.get_y() + bar.get_height()/2,
                f'R$ {bar.get_width()/1e6:.1f} M',
                va='center', # Corrected vertical alignment
                fontsize=12, color=AZUL_1)

    # Cálculo do percentual da 1ª categoria (assumindo que Eletrônicos is the first row after sorting)
    total = faturamento_por_categoria['faturamento'].sum()
    # Assuming 'Eletrônicos' is the category with the highest faturamento after sorting ascending
    eletronicos_faturamento = faturamento_por_categoria.iloc[-1]['faturamento'] # Get the last row after ascending sort
    percentual = (eletronicos_faturamento / total) * 100

    # Find the x position for the text annotation dynamically
    max_faturamento = faturamento_por_categoria['faturamento'].max()
    text_x_position = max_faturamento * 0.5 # Adjust the multiplier as needed to position the text correctly


    # Texto explicativo dentro do plot
    ax.text(text_x_position, 5, # Adjusted y position to avoid overlap
            f"$\\bf{percentual:.2f}$% das vendas correspondem apenas as vendas de Eletrônicos.\n"
            "Isso é equivalente a quase o dobro em relação a todas as outras categorias\n"
            "presentes no portfólio da Zoop.",
            fontsize=12, color=AZUL_3,
            ha='left')

    # Layout ajustado
    plt.tight_layout()
    return plt

def vendas_tri_metodo():
  # Converter a coluna 'data' para o formato datetime, se ainda não estiver
  df['data'] = pd.to_datetime(df['data'])

  # Criar uma nova coluna para identificar o trimestre
  df['trimestre'] = df['data'].dt.to_period('Q')

  # Agrupar por trimestre e método de pagamento, somando o faturamento
  vendas_por_trimestre = df.groupby(['trimestre', 'metodo_pagamento'])['faturamento'].sum().unstack()

  # Cores da identidade visual (substitua pelos valores hex adequados)
  CINZA_1 = '#ebebeb'
  AZUL_1 = '#203f75'
  CINZA_5 = '#231f20'
  BRANCO = '#ffffff'
  CINZA_3 = '#cccccc'
  AQUA_1 = '#addcd4'
  AQUA_3 = '#96bfb9'

  # Paleta de cores
  palette = [BRANCO, CINZA_3, AQUA_1, AQUA_3]

  # Visualização em gráfico de barras empilhadas
  fig, ax = plt.subplots(figsize=(10, 6))
  fig.patch.set_facecolor(CINZA_1)
  ax.set_facecolor(CINZA_1)

  vendas_por_trimestre.plot(kind='bar', stacked=True, figsize=(10, 6), color=palette, ax=ax)

  # Título
  ax.set_title('Vendas por Trimestre e Método de Pagamento em 2023', fontsize=18, color=CINZA_5)

  # Rótulos do eixo X
  ax.set_xticklabels(vendas_por_trimestre.index, rotation=0, fontsize=12, color=AZUL_1)
  ax.set_xlabel('')

  # Rótulos do eixo Y
  ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: f'R$ {x/1e6:.0f} M'))
  ax.tick_params(axis='y', labelsize=12, colors=AZUL_1)
  ax.set_ylabel('')

  # Remover bordas
  for spine in ax.spines.values():
      spine.set_visible(False)

  # Adicionar grid tracejado em y
  ax.grid(axis='y', linestyle='--', alpha=0.7, zorder=0) # zorder para colocar atrás das barras
  ax.set_axisbelow(True) # Garantir que o grid está atrás

  # Legenda
  ax.legend(title='Método de Pagamento')

  plt.tight_layout()

  return plt

def cashback():
  # Calcular a proporção de adesão ao cashback
  proporcao_cashback = df['cashback'].value_counts(normalize=True)

  # Cores da identidade visual (substitua pelos valores hex adequados)
  CINZA_1 = '#ebebeb'
  AZUL_1 = '#203f75'
  CINZA_5 = '#231f20'
  BRANCO = '#ffffff'
  CINZA_3 = '#cccccc'

  # Cores para as fatias (definidas no prompt)
  cores_fatias = [CINZA_3, AZUL_1] # Assumindo que 'Não' vem antes de 'Sim' na contagem

  # Criar gráfico de rosca
  fig, ax = plt.subplots(figsize=(6, 6), facecolor=CINZA_1)
  ax.set_facecolor(CINZA_1)

  wedges, texts, autotexts = ax.pie(
      proporcao_cashback,
      labels=proporcao_cashback.index,
      autopct='%1.1f%%',
      startangle=180,  # Ângulo de início
      colors=cores_fatias,
      wedgeprops=dict(width=0.4),  # Largura interna para criar o efeito de rosca
      pctdistance=0.8 # Posição das porcentagens
  )

  # Estilizar rótulos das fatias (labels)
  for text, wedge in zip(texts, wedges):
      text.set_color(wedge.get_facecolor()) # Cor igual à fatia
      text.set_fontsize(12)

  # Estilizar rótulos das porcentagens (autotexts)
  for autotext in autotexts:
      autotext.set_color(BRANCO) # Cor branca
      autotext.set_fontsize(12)

  # Título
  ax.set_title('Participam do programa de cashback?', fontsize=18, color=CINZA_5)

  # Manter forma circular
  ax.axis('equal')

  plt.tight_layout()

  return plt

def avaliacao_compra():
  # Cores da identidade visual (substitua pelos valores hex adequados)
  CINZA_1 = '#ebebeb'
  AZUL_1 = '#203f75'
  CINZA_5 = '#231f20'
  BRANCO = '#ffffff'
  CINZA_3 = '#cccccc'
  VERMELHO_1 = '#e23155'

  # Criar figura e eixos
  fig, ax = plt.subplots(figsize=(10, 6))
  fig.patch.set_facecolor(CINZA_1)
  ax.set_facecolor(CINZA_1)

  # Histograma
  ax.hist(df['avaliacao_compra'], bins=11, edgecolor='black', color=VERMELHO_1)

  # Título
  ax.set_title('Distribuição das avaliações das compras em 2023', fontsize=18, color=CINZA_5)

  # Rótulos dos eixos
  ax.set_xlabel('Avaliação', fontsize=12, color=AZUL_1)
  ax.set_ylabel('Frequência', fontsize=12, color=AZUL_1)

  # Rótulos dos ticks
  ax.tick_params(axis='x', labelsize=12, colors=AZUL_1)
  ax.tick_params(axis='y', labelsize=12, colors=AZUL_1)
  ax.set_xticks(range(0, 11)) # Definir ticks de 0 a 10

  # Remover bordas
  for spine in ax.spines.values():
      spine.set_visible(False)

  # Grid em y atrás do gráfico
  ax.grid(axis='y', linestyle='--', alpha=0.7)
  ax.set_axisbelow(True)

  # Adicionar caixa de anotação
  annotation_text = f"Média de avaliações = {df['avaliacao_compra'].mean().round(2)}\n" \
                    f"Avaliação mais comum = {df['avaliacao_compra'].value_counts().idxmax()}\n" \
                    f"Frequência da mais comum = {df['avaliacao_compra'].value_counts().max()}"

  ax.annotate(annotation_text,
              xy=(9, 2250), # Posição da ponta da seta
              xytext=(2, 1800), # Posição do texto
              fontsize=12,
              color=AZUL_1,
              bbox=dict(boxstyle="round,pad=0.5", fc=BRANCO, ec=CINZA_3), # Estilo da caixa
              arrowprops=dict(arrowstyle="->", color=AZUL_1)) # Estilo da seta

  plt.tight_layout()

  return plt

def dist_idade_sexo():
    # Cores da identidade visual (substitua pelos valores hex adequados)
    CINZA_1 = '#ebebeb'
    AZUL_1 = '#203f75'
    CINZA_5 = '#231f20'
    BRANCO = '#ffffff'
    CINZA_3 = '#cccccc'
    AQUA_1 = '#addcd4'

    # Criar figura e eixos
    fig, ax = plt.subplots(figsize=(10, 6))
    fig.patch.set_facecolor(CINZA_1)
    ax.set_facecolor(CINZA_1)

    # Visualização em boxplot com seaborn
    sns.boxplot(data=df, x='sexo_biologico', y='idade', palette=[AZUL_1, AQUA_1], ax=ax)

    # Título
    ax.set_title('Distribuição da Idade por Sexo Biológico dos(as) clientes', fontsize=18, color=CINZA_5)

    # Rótulos dos eixos
    ax.set_xlabel('') # Suprimir xlabel
    ax.set_ylabel('Idade', fontsize=12, color=AZUL_1)

    # Rótulos dos ticks
    ax.tick_params(axis='x', labelsize=12, colors=AZUL_1)
    ax.tick_params(axis='y', labelsize=12, colors=AZUL_1)

    # Remover bordas
    for spine in ax.spines.values():
        spine.set_visible(False)

    # Grid em y atrás do gráfico
    ax.grid(axis='y', linestyle='--', alpha=0.7)
    ax.set_axisbelow(True)

    # Adicionar caixas de anotação
    # Anotação para Sexo Biológico 'Feminino' (assumindo que é o primeiro na ordem alfabética ou como aparece no boxplot)
    feminino_stats = df.groupby("sexo_biologico")["idade"].agg(['min', lambda x: x.quantile(.25), 'median', lambda x: x.quantile(.75), 'max']).iloc[0]

    annotation_text_feminino = f"Mínimo = {feminino_stats['min']}\n" \
                               f"1º quartil = {feminino_stats['<lambda_0>']}\n" \
                               f"2º quartil = {feminino_stats['median']}\n" \
                               f"3º quartil = {feminino_stats['<lambda_1>']}\n" \
                               f"Máximo = {feminino_stats['max']}"

    ax.annotate(annotation_text_feminino,
                xy=(-0.1, 45), # Posição da ponta da seta
                xytext=(-0.4, 50), # Posição do texto
                fontsize=12,
                color=AZUL_1,
                bbox=dict(boxstyle="round,pad=0.5", fc=BRANCO, ec=CINZA_3),
                arrowprops=dict(arrowstyle="->", color=AZUL_1))

    # Anotação para Sexo Biológico 'Masculino'
    masculino_stats = df.groupby("sexo_biologico")["idade"].agg(['min', lambda x: x.quantile(.25), 'median', lambda x: x.quantile(.75), 'max']).iloc[1]

    annotation_text_masculino = f"Mínimo = {masculino_stats['min']}\n" \
                                f"1º quartil = {masculino_stats['<lambda_0>']}\n" \
                                f"2º quartil = {masculino_stats['median']}\n" \
                                f"3º quartil = {masculino_stats['<lambda_1>']}\n" \
                                f"Máximo = {masculino_stats['max']}"

    ax.annotate(annotation_text_masculino,
                xy=(0.9, 45), # Posição da ponta da seta
                xytext=(0.6, 50), # Posição do texto
                fontsize=12,
                color=AZUL_1,
                bbox=dict(boxstyle="round,pad=0.5", fc=BRANCO, ec=CINZA_3),
                arrowprops=dict(arrowstyle="->", color=AZUL_1))

    plt.tight_layout()
    return plt
````
